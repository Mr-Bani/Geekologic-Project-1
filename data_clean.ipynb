```python
import pandas as pd
data=pd.read_csv('delino_api_data_csv.csv')
data= data.iloc[:, [0,9]]
data
```

```python
import re
menu=data.iloc[0]['menu']
menu=re.sub(r"[\[\]]", "" , menu)
menu=re.split(', {',menu)
menu
```

```python
def dict_from_str(dict_str):    

    while True:

        try:
            dict_ = eval(dict_str)
        except NameError as e:
            key = e.message.split("'")[1]
            dict_str = dict_str.replace(key, "'{}'".format(key))
        else:
            return dict_
        


def every_res(data):
    menu_data=[]
    menu=re.sub(r"[\[\]]", "" , data['menu'])
    menu=re.split(', {',menu)
    for i in menu[1:]:
        d = dict_from_str('{' + i)
        d['id']=data['id']
        menu_data.append(d)
    return menu_data
        
```

```python
import csv
csv_data = "menu.csv"
data_headers = [
    "id","title","price","ingredient"]
for i in range(4919):
    m=data.iloc[i]
    l=every_res(m)
    with open(csv_data , 'a', newline='', encoding="utf-8") as ff :
        writer = csv.DictWriter(ff, data_headers)
        writer.writerows(l)
```
